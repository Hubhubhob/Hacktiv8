# Menggabungkan beberapa grafik plot, 1 figure

dalam 1 figure kita bisa membuat beberapa grafik plot digunakan function subplot (m,n,i) subplot(m,n,i)
- Parameter m dan n : sebagi dimensi dari grafikk plot
- Parameter i : sebagai urutan dari grafik plot
    contoh
    - 4 Grafik secara vertical 
        - subplott(4,1,i) akan ada 4 grafik dalamm 4 kolom 
    - 4 Grafiksecara horizontal 2 baris
        - subplot(2,2,i) akan ada 2 grafik di baris 1 dan 2 grafik dibawahnya
        
2 Grafik Horizontal
   - subplot(1,2,i) akan ada 2 grafik dalam 1 baris
    

```python
x = np.array([1,2,4,5,6])
y0 = x**2
y1 = x**3
```

```python
plt.subplot(1,2,1) # 1 x 2 plot 1
plt.plot(x,y0,"r-")
plt.title("Judul Plot Y0")
plt.xlabel("sumbu x")
plt.ylabel("Sumbu Y")
plt.legend(['Y0=x^2'])

plt.subplot(1,2,2)# 1 x 2 plot 2
plt.plot(x,y0,"r-")
plt.title("Judul Plot Y0")
plt.xlabel("sumbu x")
plt.ylabel("Sumbu Y")
plt.legend(['Y0=x^3'])

plt.suptitle('2 Grafik Plot')
plt.show()
```

```python
x = np.array([1,2,4,6])
y0 = x **2
y1 = x **3
plt.subplot(2,1,1) # 1 x 2 plot 1
plt.plot(x,y0,"r-")
plt.title("Judul Plot Y0")
plt.xlabel("sumbu x")
plt.ylabel("Sumbu Y")
plt.legend(['Y0=x^2'])

plt.subplot(2,1,2)# 1 x 2 plot 2
plt.plot(x,y0,"r-")
plt.title("Judul Plot Y0")
plt.xlabel("sumbu x")
plt.ylabel("Sumbu Y")
plt.legend(['Y0=x^3'])

plt.suptitle('2 Grafik Plot')
plt.show()
```

```python
df_can
```

```python

df_continent = df_can.groupby('Continent',axis=0).sum()

print(type(df_can.groupby('Continent',axis=0)))

df_continent.head()
```

```python
df_continent['Total'].plot(kind='pie',
                           figsize=(12,6),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True,
                          )
plt.title('Imigration to canada by continent [1980-2013]')
plt.axis('equal')
plt.show()
```

```python
colors_list = ["Gold","yellowgreen","lightblue","lightskyblue","lightskyblue","Pink"]
explode_list = [0.1,0,0,0,0.1,0.1]

df_continent["Total"].plot(kind="pie",
                           figsize=(15,6),
                           autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=colors_list,
                            explode=explode_list)# ngepop up 3
plt.title('Immigration to Canada by continent [1980-2013]',y=1.12)

plt.axis('equal')

plt.legend(labels=df_continent.index,loc='upper left')

plt.show()


```
