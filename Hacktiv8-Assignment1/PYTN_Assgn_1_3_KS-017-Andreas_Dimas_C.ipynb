# Area Plot 
Kategori kejahatan besar di London tahun 2008 - 2016

```python
major.plot(kind='area',
           alpha=0.35,
           stacked=False,
           figsize=(20, 10),
          )

plt.title('Area Plot Kejahatan Besar di London Tahun 2008 - 2016')
plt.ylabel('Total Kejahatan') 
plt.xlabel('Tahun') 

plt.show()
```

# Histograms
Kejahatan besar di London yang paling banyak dan paling sedikit dari tahun 2008 - 2016 dalam bentuk histogram!

```python
count, bin_edges = np.histogram(major, 15)
major.plot(kind ='hist',
figsize=(20, 10),
bins=15,
alpha=0.6,
xticks=bin_edges,
)

plt.title('Kejahatan Besar di London Tahun 2008 - 2016') 
plt.ylabel('Total Tahun') 
plt.xlabel('Total Kejahatan')

plt.show()
```

Kejahatan besar paling banyak di London tahun 2008 - 2016 adalah "Other Notifable Offences ". Sedangkan kejahatan besar paling sedikit adalah "Sexual Offences".

# Pie Charts

```python
totalcrimes = df_can['Year'].value_counts()
years = pd.DataFrame(data=totalcrimes.index, columns=["Year"])
years['Values'] = totalcrimes.values
```

```python
fig = px.pie(years, values='Values', names='Year', color_discrete_sequence=px.colors.qualitative.Pastel1)
fig.show()
```

```python
colors_list = ['darkslateblue','lightsteelblue','orange','moccasin','forestgreen','darkseagreen','darkorchid','plum','rosybrown']
explode_list = [0.1, 0, 0, 0, 0, 0, 0, 0.2, 0.4]
dfpie = df_can.groupby(['Crime Category']).sum()['Value'].sort_values(ascending=False)
dfpie.plot(kind='pie',
           figsize=(15, 6),
           autopct='%1.1f%%',
           startangle=90,
           shadow=True,
           labels=None, 
           colors=colors_list,
           pctdistance=1.12,
           explode=explode_list
          )
plt.title('Persentase Kejahatan Besar di London Tahun 2008 - 2016', y=1.12)
plt.ylabel('')
plt.axis('equal')

plt.legend(labels=dfpie.index, loc='upper right')

plt.show()
```

# Word Clouds

```python
from wordcloud import WordCloud, STOPWORDS
stopwords=set(STOPWORDS)
```

```python
dataregion = ' '.join(df_can.Region)
```

```python
region_wc = WordCloud(
background_color='white',
max_words=100,
stopwords=stopwords
)

# generate the word cloud
region_wc.generate(dataregion)
```

```python
plt.imshow(region_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
